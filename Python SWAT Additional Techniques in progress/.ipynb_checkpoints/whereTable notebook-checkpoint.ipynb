# DEMO: Using Additional CAS Actions

### 1. Import Packages and Connect to the CAS Server

Visit the documentation for the SWAT [(SAS Scripting Wrapper for Analytics Transfer)](https://sassoftware.github.io/python-swat/index.html) package.

```python
## Import packages
import swat
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

## Set options
pd.set_option('max_columns', None)

## Connect to CAS
conn = swat.CAS('server.demo.sas.com', 30571, 'student', 'Metadata0', name='py04d02')

## Function to load the loans_raw table into memory if necessary
def loadloans():
    conn.loadTable(path='loans_raw.sashdat', caslib='casuser',
                   casOut={'name':'loans_raw',
                           'caslib':'casuser',
                           'promote':True})
```

### 2. Explore Available CAS Tables and Data Source Files

a. Use the **tableInfo** action to view all available in-memory tables in the casuser caslib. If the **loans_raw** CAS table is not available, uncomment the statement and execute the loadloans function.

```python
#loadloans()
conn.tableInfo(caslib='casuser')
```

b. Reference the **loans_raw** CAS table where **Category** equals *Mortgage*. Then preview the table.

```python
mTbl = conn.CASTable('loans_raw', caslib='casuser', where = "Category = 'Mortgage'")
display(mTbl)
mTbl.head()
```

### 3. Simple Action Set


a. You can use the [simple.correlation](https://go.documentation.sas.com/doc/en/pgmsascdc/v_018/casanpg/n11jidodvxk3tkn1sy8636iha1po.htm#p0jytlrw1nn19tn1782gu5c2cv43) to generate a matrix of Pearson correlation coefficients for two or more input columns. By default, univariate descriptive statistics are also generated for the analysis variables. You can disable the univariate descriptive statistics if you do not need them.

```python
colNames = ['Age', 'EmpLength', 'Amount', 'InterestRate', 'LoanLength']
mTbl.correlation(inputs = colNames)

## Alternate version
# mTbl.vars = ['Age', 'EmpLength', 'Amount', 'InterestRate', 'LoanLength']
# mTblcorrelation()
```

b. The SWAT package also provides the corr method.

```python
mTbl[colNames].corr()
```

c. Use the [simple.topk](https://go.documentation.sas.com/doc/en/pgmsascdc/v_018/casanpg/cas-simple-topk.htm) action to return the top-K and bottom-K distinct values of each column included in the list. In this example the top *5* and bottom *5* mortgages will be returned. The inputs parameter specifies the column or columns to analyze, and raw parameter specifies to use the raw values and avoid masking values with any formats. The action returns two **SASDataFrames**. The first shows the top 5 and bottom 5 **InterestRates**, and the second shows the number of unique values in the column.

```python
mTbl.topk(inputs = 'InterestRate',
          topk = 5,
          bottomk = 5,
          raw = True) # <-avoid masking by auto formatting values
```

d. You can also use an aggregator within the topk action to find the top and bottom number based on an aggreagation. Here we will specify the mean aggreagator to aggregate the **Amount** by each **LoanGrade**, then return the **LoanGrade** with the highest aggreagation, and the lowest aggregation. 

```python
mTbl.topk(inputs = 'LoanGrade',  ## <--group for the aggregator
          weight = 'Amount',     ## <--specify the column to aggregate
          aggregator = 'MEAN',   ## <--how to aggregate
          raw = True)
```

### 4. Percentile Action Set

a. Load and view the percentile action set.

```python
conn.loadActionSet('percentile')
conn.percentile?
```

b. Group the mortgage loans by **LoanGrade**, then execute the [percentile.boxplot](https://go.documentation.sas.com/doc/en/pgmsascdc/v_018/casanpg/cas-percentile-boxplot.htm) action to calculate quantiles, high and low whiskers, and outliers. Since the table is grouped you must use the concat_bygroups method to concatenate each individual by group. Then call the *BoxPlot* key to return the DataFrame.

**NOTE**: If you do not specify the action set, the boxplot plotting method will be called.

```python
mTbl.groupBy = ['LoanGrade']

mTbl.percentile.boxplot(inputs = 'InterestRate').concat_bygroups()['BoxPlot']
```

e. The [percentile.percentile](https://go.documentation.sas.com/doc/en/pgmsascdc/v_018/casanpg/cas-percentile-percentile.htm) action calculate quantiles and percentiles. Here the groupby parameter is deleted, then the percentile action is executed on the **Amount** and **InterestRate** columns. Notice the action returns a **CASResults** object with a single **SASDataFrame** with percentiles for each column.

```python
del mTbl.groupby
mTbl.percentile(inputs = ['Amount', 'InterestRate'])
```

### 5. DataPreprocess

a.Load and view the dataPreprocess action set.

```python
conn.loadActionSet('dataPreprocess')
conn.dataPreprocess?
```

b. Execute thie [dataPreprocess.histogram](https://go.documentation.sas.com/doc/en/pgmsascdc/v_018/casanpg/cas-datapreprocess-histogram.htm) action to generate histogram bins and simple bin-based statistics for numeric columns. Store the **SASDataFrame** from the results of the histogram action by calling the *BinDetails* key in the **CASResults** object.

```python
histDf = mTbl.histogram(inputs = ['InterestRate'],
                        requestPackages = [{'nbins':10, 
                                            'binStart':0,
                                            'niceBinning':False}])['BinDetails']

histDf
```

c Using the **histDf** variable, you can use the **MidPoint** and **Percent** columns to plot a bar chart to visualize the histogram results.

**NOTE**: The histogram action allows you to summarize extremely large tables, then you can use the summary of the action to visualize the results.

```python
ax = plt.figure(figsize=(10,5))
ax = sns.barplot(data = histDf, x = 'MidPoint', y='Percent', color='blue')
ax.set(title='Histogram of Mortgage Interest Rates')
```

### 6. datashaping Action Set

a. a.Load and view the dataShaping action set.

```python
conn.loadActionSet('dataShaping')
conn.dataShaping?
```

b. To work with the dataShaping action we will create a small table as a demonstration. First, load the *customers_raw.csv* file into memory and use only the selected columns. Then, reference the new CAS table in the variable **customers**. Lastly, create a **SASDataFrame** with 5 rows in the variable **df**.

```python
## Load the customers_raw.csv file into memory with the specified parameters
conn.loadTable(path='customers_raw.csv', caslib='casuser', 
               vars = ['ID','LoanCreditCard', 'SavingsAcct', 'CheckingAcct'],
               casOut={'replace':True})

## Reference the CAS table
customers = conn.CASTable('customers_raw', caslib = 'casuser')

## Create a small wide DataFrame
df = customers.sort_values('ID').head()
df
```

c. Upload the **SASDataFrame** to a CAS table using the upload action. Name the new CAS table **wideTest**.

```python
conn.upload(df, casOut={'name':'wideTest', 'replace':True})
```

d. Transform the wide CAS table to a long table using the data [dataShaping.wideToLong](https://go.documentation.sas.com/doc/en/pgmsascdc/v_018/casanpg/cas-datashaping-widetolong.htm) action.

```python
inputTbl = {'name':'wideTest', 'caslib':'casuser'}
outputTbl = dict(name = 'LongTest', caslib = 'casuser', replace = True)

conn.wideToLong(inputTbl,
                id = 'ID', 
                inputs = ['LoanCreditCard','SavingsAcct','CheckingAcct'],
                variableName = 'AccountType',
                valueName='AccountExists',
                casOut=outputTbl)
```

d. Preview the new table

```python
testTbl = conn.CASTable('LongTest', caslib = 'casuser')
testTbl.head(15)
```

### 7. Advanced Filtering Using a Table (I think this is a lot. Probably remove)
**Scenario**: Create a new CAS table with a list of all customers who have cancelled at least one account, then use that tablet to filter our full list of customers.

a. Use the **loans_raw** table to find all customers who have cancelled an account and create a new CAS table that contains the **ID** of those customers. First, create a new variable named **loansTbl** that references the **loans_raw** CAS table, use the where parameter to find all cancelled accounts, then only output the **ID** column. Use the **loansTbl** object and execute the copyTable action to create a new table named **cancelledloans** in the **Casuser** caslib. 
Reference the newly created CAS table **cancelledLoans** in the variable **cancelled**. Then preview the table and print the number of rows in the new table.

```python
## Create a CASTable object that finds all cancelled accounts
loansTbl = conn.CASTable('loans_raw', caslib='casuser', 
                         where = 'Cancelled = 1',
                         vars = ['ID'])

## Create a new CAS table with the IDs of all cancelled accounts
loansTbl.copyTable(casOut = {'name':'cancelledLoans', 
                             'caslib':'casuser', 
                             'replace':True})

## Reference the filter table
cancelled = conn.CASTable('cancelledLoans', caslib='casuser')

## Preview the new table
display(cancelled.head())
print('Number of rows:{:,}'.format(cancelled.shape[0]))
```

b. Display the available data source files in the **Casuser** caslib. Then load the **customers_raw.csv** file into memory and execute the tableInfo action to view the available tables in memory.

```python
## View available data source files
display(conn.fileInfo(caslib = 'casuser'))

## Load the customres_raw.csv file
conn.loadTable(path='customers_raw.csv', caslib='casuser', 
               casOut={'replace':True})

## Confirm the file was loaded into memory
display(conn.tableInfo(caslib = 'casuser'))
```

c. Reference the **customers_raw** CAS table in the variable **customers**. Preview the table by executing the shape attribute and head method. Notice the **customers_raw** table contains over 10 million rows, and contains information for each customer.

```python
customers = conn.CASTable('customers_raw', caslib='casuser')

print('Number of rows:{:,}'.format(customers.shape[0]))
customers.head()
```

d. Add the whereTable parameter to the **customers** CASTable object. In the whereTable parameter reference the **cancalledLoans** CAS table, and the column **ID**. The whereTable parameter will filter a table using another CAS table. You can almost think of this as using a subquery in SQL, or using the IN operator with a list.

```python
customers.whereTable = {'name':'cancelledLoans', 'vars':'ID'}
customers
```

e. Create a new table with a list of all customers who have cancelled at least one account. Name the table **cancelledCustomerInfo**, place it in the **Casuser** caslib, and add a label to the table. Then execute the tableInfo action to view the **Name**, **Label**, **Rows**, and **Columns** columns of all available in-memory tables. Notice the new table has about 1.4 million rows.

```python
customers.copyTable(casOut = {'name':'cancelledCustomerInfo', 
                              'caslib':'casuser',
                              'label':'Table with all customers who have cancelled at least one account',
                              'replace':True})

conn.tableInfo(caslib = 'casuser')['TableInfo'].loc[:,['Name', 'Label', 'Rows','Columns']]
```
